```python
#drop observations with less than 27 plate appearances

bs_15.drop(bs_15[bs_15['PA▼'] < 27].index, inplace=True)
```

```python
bs_15.columns
non_numerical_col = ['Rk','Name','Age','Tm','Lg','Pos Summary']
```

```python
# list of numerical variables
num_vars = [var for var in bs_15.columns if bs_15[var].dtypes != 'O']

print('Number of numerical variables: ', len(num_vars))

# visualise the numerical variables
bs_15[num_vars].head()
```

```python
non_numerical_var = ['Rk', 'Name', 'Tm', 'Lg', 'Pos Summary']
```

```python
#  list of discrete variables
discrete_vars = [var for var in num_vars if len(bs_15[var].unique())<20 and var not in non_numerical_var]

print('Number of discrete variables: ', len(discrete_vars))
discrete_vars
```

```python
def analyse_discrete(df, var):
    df = df.copy()
    df.groupby(var)['HR'].median().plot.bar()
    plt.title(var)
    plt.ylabel('HR')
    plt.show()
    
for var in discrete_vars:
    analyse_discrete(bs_15, var)
```

```python
# list of continuous variables
cont_vars = [var for var in num_vars if var not in discrete_vars+non_numerical_var]

print('Number of continuous variables: ', len(cont_vars))
```

```python
# let's visualise the continuous variables
bs_15[cont_vars].head()
```

```python
# Let's go ahead and analyse the distributions of these variables
def analyse_continous(df, var):
    df = df.copy()
    df[var].hist(bins=20)
    plt.ylabel('Number of HR')
    plt.xlabel(var)
    plt.title(var)
    plt.show()
    
for var in cont_vars:
    analyse_continous(bs_15, var)
```

```python
# Let's go ahead and analyse the distributions of these variables
def analyse_transformed_continous(df, var):
    df = df.copy()
    
    # log does not take negative values, so let's be careful and skip those variables
    if 0 in bs_15[var].unique():
        pass
    else:
        # log transform the variable
        df[var] = np.log(df[var])
        df[var].hist(bins=20)
        plt.ylabel('Number of houses')
        plt.xlabel(var)
        plt.title(var)
        plt.show()
    
for var in cont_vars:
    analyse_transformed_continous(bs_15, var)
```

```python

def transform_analyse_continous(df, var):
    df = df.copy()
    
    # log does not take negative values, so let's be careful and skip those variables
    if 0 in df[var].unique():
        pass
    else:
        # log transform
        df[var] = np.log(df[var])
        df['HR'] = np.log(df['HR'])
        plt.scatter(df[var], df['HR'])
        plt.ylabel('HR')
        plt.xlabel(var)
        plt.show()
    
for var in cont_vars:
    if var !='HR':
        transform_analyse_continous(bs_15, var)
```

```python
# let's explore the relationship between the home runs and the transformed variables
# with more detail
def transform_analyse_continous(df, var):
    df = df.copy()
    
    # log does not take negative values, so let's be careful and skip those variables
    if 0 in bs_15[var].unique():
        pass
    else:
        # log transform
        df[var] = np.log(df[var])
        df['HR'] = np.log(df['HR'])
        plt.scatter(df[var], df['HR'])
        plt.ylabel('HR')
        plt.xlabel(var)
        plt.show()
    
for var in cont_vars:
    if var !='HR':
        transform_analyse_continous(bs_15, var)
```

    -PA: In baseball statistics, a player is credited with a plate appearance (denoted by PA) each time he completes a turn batting. ... A very similar statistic, at bats, counts a subset of plate appearances that end under certain circumstances.
    -AB: Plate appearances, not including bases on balls, being hit by pitch, sacrifices, interference, or obstruction.
    -R: A player is awarded a run if he crosses the plate to score his team a run. When tallying runs scored, the way in which a player reached base is not considered.
    -H: when the ball comes in contact with the bat
    -RBI: when a batter makes a play that that allows a run to be scored
    -BB: Base on balls-when a batter is awarded first base after not swinging at 4 pitches that are called out of the strike zone
    -SO: Number of hits divided by number of at-bats
    -TB: total number of bases gained from a Hit. It is a weighted sum for which the weight value is 1 for a single, 2 for a double, 3 for a triple and 4 for a home run.
